```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

f1 = open("temp-d5-2b-uncov.txt", "r")
f2 = open("temp-d6-ce-poe-uncov.txt", "r")

f3 = open("temp-d5-2b-cov.txt", "r")
f4 = open("temp-d6-ce-poe-cov.txt", "r")

y1 = []
y2 = []
y3 = []
y4 = [] 

for i in f1.readlines()[2:]:
    y1.append(float(i.strip('\n'))/1000)

for i in f2.readlines()[2:]:
    y2.append(float(i.strip('\n'))/1000)

for i in f3.readlines()[2:]:
    y3.append(float(i.strip('\n'))/1000)

for i in f4.readlines()[2:]:
    y4.append(float(i.strip('\n'))/1000)
```

```python
# Generate the Xs

figure(figsize=(12, 8))
plt.plot(range(0, len(y1)*10, 10), y1, label="Micro UCSB power w/o fan, uncovered")
plt.plot(range(0, len(y3)*10, 10), y3, label="Micro UCSB power w/o fan, covered")
plt.plot(range(0, len(y2)*10, 10), y2, label="PoE with fan, uncovered")
plt.plot(range(0, len(y4)*10, 10), y4, label="PoE with fan, covered")

# Empahize the starting temperatures
plt.plot(0,y1[0],'bo')
plt.plot(0,y3[0],'o', color='orange')
plt.plot(0,y2[0],'go')
plt.plot(0,y4[0],'ro')

plt.xlabel('Time (s)')
plt.ylabel(u'Temperature (\u00B0C)')
plt.legend()
plt.show()
```
